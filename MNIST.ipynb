```python
X_train.shape #shape of training data 60000 images, each of 28 rows and 28 columns.
```

```python
X_train=X_train/255 #sclading down the data to improve accuracy
X_test = X_test/255
#Data scaling is the process of transforming numerical features into a smaller or standardized range without changing the underlying relationships between the data points.
#It changes the magnitude of values, not their meaning.
```

```python
#reshaping 28by28 matrix into 1 by 74 matrix or array, converting a multi-dimensional input into a one-dimensional vector.
X_train_flattened=X_train.reshape(len(X_train),28*28) 
X_train_flattened.shape
```

```python
X_test_flattened=X_test.reshape(len(X_test),28*28)
X_test_flattened.shape
```

```python
X_train_flattened[0]
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Input

model = Sequential([                              #Createing a model of 1 input, 1 output and 3 hidden layers.
    Input(shape=(784,)),
    Dense(128, activation='relu'),
    Dense(100, activation = 'sigmoid'),
    Dense(50, activation = 'sigmoid'),
    Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy',metrics=['accuracy']) 
x = model.fit(X_train_flattened, y_train, epochs=40)
```

```python
model.evaluate(X_test_flattened,y_test) # checking model accuracy for test dataset
```

```python
plt.matshow(X_test[0])
plt.show()
```

```python
y_pred=model.predict(X_test_flattened) # the result is the values in all 10 neurons of the output layer.
y_pred[0]
```

```python
y_pred_labels=[np.argmax(i) for i in y_pred] 
y_pred_labels[:5]
```

```python
y_test[:5]
```

```python
np.argmax(y_pred[0]) # np.argmax() returns the index of the maximum value from the y_pred
```

```python
cm=tf.math.confusion_matrix(labels = y_test, predictions=y_pred_labels) #plotting confusion matrix
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
plt.figure()
plt.plot(x.history['loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

```python
plt.figure()
plt.plot(x.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'])
plt.show()
```
